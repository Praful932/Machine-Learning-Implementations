```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import pandas as pd
from sklearn import svm
from scipy.io import loadmat
```

```python
df = loadmat('ex6data1.mat')
X = df['X']
# Convert to 1D Array
y = df['y'].squeeze()
pos = y == 1
neg = y == 0
plt.scatter(X[pos,0],X[pos,1],label='Positive',marker='+')
plt.scatter(X[neg,0,],X[neg,1],label='Negative',marker='o')
plt.legend()
```

```python
# Use Linear Support Vector Classification inbuilt from scikit library
# With C=1
sv = svm.LinearSVC(C=1,loss='hinge',max_iter=100000)
sv.fit(X,y)
sv.score(X,y)
```

```python
# With C=100
sv2 = svm.LinearSVC(C=100,max_iter=100000)
sv2.fit(X,y)
sv2.score(X,y)
```

```python
# To see Confidence Levels for C=1
confx_c1 = sv.decision_function(X)
plt.scatter(X[:,0],X[:,1],c = confx_c1,cmap='seismic')
plt.title("SVM(C=1) Decision Confidence")
```

```python
confx_c100 = sv2.decision_function(X)
plt.scatter(X[:,0],X[:,1],c = confx_c100,cmap='seismic')
plt.title("SVM(C=100) Decision Confidence")
# Here SVM Classifies every eg. correctly with c=100, but it is not a 
# natural fit for the given data
```

```python
def GaussianKernel(x1, x2, sigma):
    tosum = (x1-x2) ** 2
    return np.exp(-sum(tosum)/(2*(sigma**2)))
```

```python
sigma = 2
x1 = np.array([1,2,1])
x2 = np.array([0,4,-1])
print(GaussianKernel(x1,x2,sigma))
```

```python
df = loadmat('ex6data2.mat')
DataSet2 = pd.DataFrame(data=df['X'],columns=['X1','X2'])
DataSet2['y'] = df['y']
DataSet2.head()
```

```python
# Visualize Dataset
pos=df['y']==1
neg=df['y']==0
plt.scatter(DataSet2[pos]['X1'],DataSet2[pos]['X2'],label="Positive")
plt.scatter(DataSet2[neg]['X1'],DataSet2[neg]['X2'],label="Negative")
plt.legend()
```

```python
# Train SVM with RBF Kernel
rbf_svc = svm.SVC(kernel='rbf',gamma=15,probability=True)
rbf_svc.fit(DataSet2[['X1','X2']],DataSet2['y'])
plt.scatter(DataSet2['X1'],DataSet2['X2'],c=rbf_svc.predict_proba(DataSet2[['X1','X2']])[:,0],cmap='magma_r')
confidence = rbf_svc.score(DataSet2[['X1','X2']],DataSet2['y'])
print(f"Performance :{confidence}")
```

```python
# DataSet3
df = loadmat('ex6data3.mat')
DataSet3Train = pd.DataFrame(data=df['X'],columns=['X1','X2'])
DataSet3Train['y'] = df['y']
DataSet3Val = pd.DataFrame(data=df['Xval'],columns=['X1','X2'])
DataSet3Val['yval'] = df['yval']
# Visualize DataSet
pos = df['y']==1
neg = df['y'] == 0
plt.scatter(DataSet3Train[pos]['X1'],DataSet3Train[pos]['X2'],label='Positive')
plt.scatter(DataSet3Train[neg]['X1'],DataSet3Train[neg]['X2'],label='Negative')
plt.legend()
```

```python
# Find Best C and Gamma Value
ctest = [0.01,0.03,0.1,0.3,1,3,10,30]
gammatest = [0.01,0.03,0.1,0.3,1,3,10,30]
BestScore = 0
BestParams ={'C':None,'gamma':None}
for i in ctest:
    for j in gammatest:
        svmodel = svm.SVC(C=i,gamma=j)
        svmodel.fit(DataSet3Train[['X1','X2']],DataSet3Train['y'])
        score = svmodel.score(DataSet3Val[['X1','X2']],DataSet3Val['yval'])
        if(score>BestScore):
            BestScore = score
            BestParams['C'] = i
            BestParams['gamma'] = j
BestScore, BestParams
```

```python
# Train a Spam Classifier
df = loadmat('spamTrain.mat')
X = df['X']
y = df['y'].flatten()
svm_spam = svm.LinearSVC(C=0.1)
svm_spam.fit(X,y)
# Training Accuracy
svm_spam.score(X,y) * 100
```

```python
df = loadmat('spamTest.mat')
Xtest = df['Xtest']
ytest = df['ytest'].flatten()
# Test Accuracy
svm_spam.score(Xtest,ytest) * 100
```
