```python
def sigmoid(X):
    return 1/(1 + np.exp(-X))
```

```python
def predict(X, Theta1,Theta2):
    m = X.shape[0]
    num_labels = Theta2.shape[0]
    p = np.zeros((X.shape[0], 1))
    a1 = np.hstack((np.ones((m,1)), X))
    z2 = a1 @ Theta1.T
    a2 = sigmoid(z2)
    a2 = np.hstack((np.ones((m,1)), a2))
    z3 = a2 @ Theta2.T
    ghx = sigmoid(z3)
    pred = np.argmax(ghx,axis=1)
    return pred
```

```python
sel = np.random.randint(0,m)
plt.imshow(X[sel,:].reshape(20,20, order = "F"),cmap="magma")
plt.axis("off")
p= predict((X[sel,:][:,np.newaxis]).T, Theta1, Theta2)
p[0] = p[0]+1 if p[0]!=9 else 0
print(f"The Predicted number is :{p[0]}")
```
